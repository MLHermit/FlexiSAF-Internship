plotting the total profit column with lineplot

```python
plt.plot(path['month_number'], path['total_profit'], marker='o', color='b')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit by Month') 
plt.show()
```

The month of November produced the most profit of over 400,000 of value

```python
fig, axs = plt.subplots(2, 1)

# Bathing soap sales
axs[0].plot(path['month_number'], path['bathingsoap'], marker='o', color='green')
axs[0].set_title('Bathing Soap Sales per Month')
axs[0].set_ylabel('Units Sold')
axs[0].grid(True)

# Facewash sales
axs[1].plot(path['month_number'], path['facewash'], marker='o', color='purple')
axs[1].set_title('Facewash Sales per Month')
axs[1].set_xlabel('Month Number')
axs[1].set_ylabel('Units Sold')
axs[1].grid(True)
plt.tight_layout() #for spacing between subplots
plt.show()
```
