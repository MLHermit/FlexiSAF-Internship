Ex 2. Calculating summary statistics for each numerical feature

```python
for col in nyc_taxi.columns:
    #print((nyc_taxi[col]).dtype)
    if (nyc_taxi[col].dtype) not in ['datetime64[us]', 'object']:
        print(np.mean(nyc_taxi[col]), '\t', np.median(nyc_taxi[col]), '\t', np.std(nyc_taxi[col]))
       #print(nyc_taxi.col.mean, '\n', col.median(), '\n', col.mode(), '\n', col.std())
```
